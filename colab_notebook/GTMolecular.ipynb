Required setup for automating bwa run:

```python
from pathlib import Path
import glob

files = glob.glob("*.fastq")

index = "GCA_000001405.15_GRCh38_full_analysis_set.fna"
```

Running bwa to align reads to GRCh38 bwa indexes:

```python
for fq in files:
    sample = Path(fq).stem
    rg = f"@RG\\tID:{sample}\\tLB:{sample}\\tPL:ILLUMINA\\tSM:{sample}"

    !bwa mem -t 8 -M -R '{rg}' {index} {fq} > {sample}.sam
```
